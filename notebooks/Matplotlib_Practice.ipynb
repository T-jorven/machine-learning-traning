```python
# Create an array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X

X = np.linspace(0, 100, 100)
Y = X ** 2

# Create a plot using plt.subplots() and plot X versus X^2 (X squared)

fig, ax = plt.subplots()
ax.plot(X, Y, color="blue", label="X squared")

ax.set_xlabel("X")
ax.set_ylabel("X")
ax.set_title("Plot of X vs X")
ax.legend()

fig.savefig("x_squared_plot.png", dpi=300)

plt.show();
```

```python
#We'll start with scatter plots.

# Create a scatter plot of X versus the exponential of X (np.exp(X))

X = np.linspace(0, 100, 100)

# Create a scatter plot of X versus np.sin(X)

# --- Scatter: X vs exp(X) ---
fig1, ax1 = plt.subplots()
ax1.scatter(X, np.exp(X))
ax1.set_xlabel("X")
ax1.set_ylabel("exp(X)")
ax1.set_title("Scatter: X vs exp(X)")

# --- Scatter: X vs sin(X) ---
fig2, ax2 = plt.subplots()
ax2.scatter(X, np.sin(X))
ax2.set_xlabel("X")
ax2.set_ylabel("sin(X)")
ax2.set_title("Scatter: X vs sin(X)")

plt.show()
```

```python
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
# Plot X versus X/2 on the top left axes
# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
# Plot a bar graph of the favourite food keys and values on the bottom left subplot
# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot

# Data
X = np.linspace(0, 10, 100)

favorite_foods = {
    "Sushi": 12.5,
    "Lasagna": 9.0,
    "Mango": 3.2
}
foods = list(favorite_foods.keys())
prices = list(favorite_foods.values())

rng = np.random.default_rng()

# Figure with 2 rows x 2 cols
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# Top-left: X vs X/2
axes[0, 0].plot(X, X / 2)
axes[0, 0].set_title("X vs X/2")
axes[0, 0].set_xlabel("X")
axes[0, 0].set_ylabel("X/2")

# Top-right: Scatter of 10 random numbers
axes[0, 1].scatter(rng.random(10), rng.random(10))
axes[0, 1].set_title("Random Scatter (10 pts)")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("y")

# Bottom-left: Bar of favorite foods
axes[1, 0].bar(foods, prices)
axes[1, 0].set_title("Favorite Foods")
axes[1, 0].set_xlabel("Food")
axes[1, 0].set_ylabel("Price ()")

# Bottom-right: Histogram of 1000 normal numbers
axes[1, 1].hist(rng.normal(0, 1, 1000), bins=30)
axes[1, 1].set_title("Normal(0,1) Histogram")
axes[1, 1].set_xlabel("Value")
axes[1, 1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Import the '../data/car-sales.csv' into a DataFame called car_sales and view

df = pd.read_csv("../data/xportedcarsales.csv")

# Try to plot the 'Price' column using the plot() function

# Plot the 'Price' column
df["Price"].plot()

plt.title("Car Prices")
plt.xlabel("Index")
plt.ylabel("Price")
plt.show()
```

```python
# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
# View the car_sales DataFrame


# 1) Load CSV (first column is an index)
df = pd.read_csv("../data/xportedcarsales.csv", index_col=0)

# 2) Clean 'Price': remove $, commas, and cents (last two digits), then to int
#    e.g. "$4,000.00" -> "400000" -> remove last 2 digits -> "4000" -> 400
df["Price"] = (
    df["Price"].str.replace(r"[^0-9]", "", regex=True)  # keep only digits
                    .str[:-2]                           # drop cents
                    .astype(int)
)

# 3) Add cumulative total of 'Price'
df["Total Sales"] = df["Price"].cumsum()

# 4) Add successive sale dates starting from today (Europe/Berlin)
start = pd.Timestamp.now(tz="Europe/Berlin").normalize()
df["Sale Date"] = pd.date_range(start=start, periods=len(df), freq="D")

# 5) View the DataFrame
df
```

```python
# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column

df.plot(x="Sale Date", y="Total Sales", kind="line", marker="o")

plt.title("Total Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales ($)")
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Price' column to the integers
# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function


# 1) Force to string
# 2) Remove all non-digits ($, commas, dot)
# 3) Drop the last 2 digits (cents)
# 4) Convert to nullable integer

price_digits = df["Price"].astype(str).str.replace(r"\D", "", regex=True)
df["Price"] = pd.to_numeric(price_digits, errors="coerce") // 100
df["Price"] = df["Price"].astype("Int64")  # keep as int (nullable)

# Ensure Odometer is numeric too (just in case)
df["Odometer (KM)"] = pd.to_numeric(df["Odometer (KM)"], errors="coerce")

# --- Scatter plot: Odometer vs Price ---
df.plot(x="Odometer (KM)", y="Price", kind="scatter")
plt.title("Car Price vs Odometer Reading")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price ($)")
plt.show()
```

```python
# Create a bar graph of the 'Make' and 'Odometer (KM)' columns in the car_sales DataFrame
# Create a histogram of the 'Odometer (KM)' column
# Create a histogram of the 'Price' column with 20 bins 


import pandas as pd
import matplotlib.pyplot as plt

# --- Load your CSV (adjust name if needed) ---
df = pd.read_csv("../data/xportedcarsales.csv", index_col=0)

# --- Clean columns ---
# Odometer numeric
df["Odometer (KM)"] = pd.to_numeric(df["Odometer (KM)"], errors="coerce")

# Price: strip $, commas, cents -> integers
price_digits = df["Price"].astype(str).str.replace(r"\D", "", regex=True)
df["Price"] = (pd.to_numeric(price_digits, errors="coerce") // 100).astype("Int64")

# 1) Bar graph of 'Make' and 'Odometer (KM)' (average odometer by make)
df.groupby("Make")["Odometer (KM)"].mean().plot(kind="bar")
plt.title("Average Odometer (KM) by Make")
plt.xlabel("Make")
plt.ylabel("Average Odometer (KM)")
plt.tight_layout()
plt.show()

# 2) Histogram of 'Odometer (KM)'
df["Odometer (KM)"].plot(kind="hist", bins=15)
plt.title("Histogram of Odometer (KM)")
plt.xlabel("Odometer (KM)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# 3) Histogram of 'Price' with 20 bins
df["Price"].dropna().astype(int).plot(kind="hist", bins=20)
plt.title("Histogram of Price ($)")
plt.xlabel("Price ($)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Replicate the above plot in whichever way you see fit
# Note: The method below is only one way of doing it, yours might be
# slightly different
# Create DataFrame with patients over 50 years old
# Create the plot
# Plot the data
# Customize the plot
# Add a meanline


# Load the file
heart_disease = pd.read_csv("../data/heart-disease.csv")

# Keep patients > 50
over_50 = heart_disease[heart_disease["age"] > 50].copy()

# Plot (replicates the figure)
fig, ax = plt.subplots(figsize=(12, 5))
for cls, grp in over_50.groupby("target"):
    ax.scatter( grp["age"],
        grp["chol"],
        c=np.full(len(grp), cls),  # 0s or 1s
        cmap="winter",
        vmin=0, vmax=1,
        label=str(cls))
    plt.style.use("fast");

ax.set_title("Heart Disease and Cholesterol Levels")
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
ax.legend(title="Target")

# Mean line
mean_chol = over_50["chol"].mean()
ax.axhline(mean_chol, linestyle="--", color="red")
ax.text(over_50["age"].min(), mean_chol + 3, f"Mean = {mean_chol:.0f}", va="bottom")

plt.tight_layout()
plt.savefig("my_figure.png", dpi=300, bbox_inches="tight")
plt.show();

```
